```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

Choosing the page to extract data from

```python
webpage=requests.get("https://www.eventbrite.com/d/united-states/all-events/?page=4").text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
#print(soup.prettify())
```

The required data was in listed format inside the page.

```python
events= soup.find_all('li',class_='search-main-content__events-list-item search-main-content__events-list-item__mobile')
```

Taking names,links of event from the page. 

```python
name=[]
links=[]
for i in events:
    name.append(i.find('h2',class_="Typography_root__4bejd").text)
    links.append(i.find('a',class_="event-card-link")['href'])
    
```

```python
name
```

```python
links
```

Creating a dataframe of the event datas. 

```python
date_time=[]
location=[]
price=[]
about=[]
for link in links:
    event_url=requests.get(link).text
    so=BeautifulSoup(event_url,'lxml')
    event=so.find_all('div',class_='eds-structure__body')
    for j in  event:
        date_time.append(j.find('span',class_="date-info__full-datetime").text)
        about.append(j.find('div',class_="has-user-generated-content").text)
        try:
            price.append(j.find('div',class_="conversion-bar__panel-info").text)
        except:
            price.append(np.nan)
        try:
            location.append(j.find('p',class_="location-info__address-text").text)
        except:
            location.append(np.nan)
    
    
df =pd.DataFrame({'Name':name,'Price':price,'Date_time':date_time,"Location":location,"About":about})
```

```python
df
```

Creating a dataframe for the events in USA from all the 500 pages of event listed

```python
data_list = []

for a in range(1, 501):
    webpage = requests.get("https://www.eventbrite.com/d/united-states/all-events/?page={}".format(a)).text
    soup = BeautifulSoup(webpage, 'lxml')
    events = soup.find_all('li', class_='search-main-content__events-list-item')
    
    for event in events:
        name = event.find('h2', class_="Typography_root__4bejd").text
        link = event.find('a', class_="event-card-link")['href']
        
        event_url = requests.get(link).text
        so = BeautifulSoup(event_url, 'lxml')
        event_info = so.find('div', class_='eds-structure__body')
        
        try:
            about = event_info.find('div', class_="has-user-generated-content").text
        except AttributeError:
            about = np.nan
               
        try:
            date_time = event_info.find('span', class_="date-info__full-datetime").text
        except AttributeError:
            date_time = np.nan
        
        try:
            price = event_info.find('div', class_="conversion-bar__panel-info").text
        except AttributeError:
            price = np.nan
        
        try:
            location = event_info.find('p', class_="location-info__address-text").text
        except AttributeError:
            location = np.nan
        
        # Append data as dictionaries to the list
        data_list.append({'Name': name, 'Price': price, 'Date_time': date_time, 'Location': location, 'About': about})

# Create a DataFrame from the list of dictionaries
final = pd.DataFrame(data_list)

```
