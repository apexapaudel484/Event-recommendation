```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
webpage=requests.get("https://www.eventbrite.com/d/nepal/all-events/?page=1").text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
events= soup.find_all('li',class_='search-main-content__events-list-item search-main-content__events-list-item__mobile')
for i in events:
    print(i.text)
```

```python
name=[]
date_time=[]
location=[]
price=[]
links=[]
for i in events:
    name.append(i.find('h2',class_="Typography_root__4bejd").text)
    links.append(i.find('a',class_="event-card-link")['href'])
    
```
